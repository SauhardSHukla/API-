### Oragnized Data Fetching


```python
import requests
import json
import time
import pandas as pd
import matplotlib as plot
```

```python
get_url = 'https://dummyjson.com/carts'
request_data = requests.get(get_url)
request_data.json()
```

```python
row=[]
for cart_info in request_data.json()['carts']:
    cart_id  = cart_info['id']
    # print(cart_info)
    print(f'Getting info of the cart id {cart_id}')

    for product_info in cart_info['products']:
            # print(product_info)
            
                # product_info_data =  product_info['id']
                row_info=[]

                # pid  = product_info['id'] Not need 

                title = product_info['title']
                price = product_info['price']
                quantity = product_info['quantity']

                # Append all the data in the row_info list
                row_info.append(cart_id)
                row_info.append(title)
                row_info.append(price)
                row_info.append(price)

                row.append(row_info)
    
```

```python
df=pd.DataFrame(row,columns=['cart_id','title','price','quantity'])
df
```

1. Average price of items in the cart
2. Average qunatity of items in the cart
3. Product most freq. bought

```python
#We will group the data here

df.groupby(['cart_id'],as_index=False).agg({'price':'mean'})
```

```python
# Plot to show the all locations of the dot to get the visual idea of the Data

df.groupby(['cart_id'],as_index=False).agg({'price':'mean'}).plot(marker ='o',linestyle='--')
```
